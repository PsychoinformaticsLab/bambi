# Notebook to track progress on ADVI implementation

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc3 as pm

import bambi as bmb
```

## Generate Random Data

```python
a = np.random.normal(5, 2, 1000)
b = np.random.normal(8, 2.5, 1000)

df = pd.DataFrame(
    {
        'Group': ['a'] * 1000 + ['b'] * 1000,
        'Val': np.hstack([a, b])
    }
)
```

## Create Model

```python
model = bmb.Model(df)
```

```python
results = model.fit('Val ~ Group', backend='pymc3', run=True, method='advi')
```

## Plotting

```python
results.plot()
```
